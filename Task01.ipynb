```python
import random
import pandas as pd
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
unique_value = data['whoAmI'].unique()
unique_value
```

```python
one_hot_df = data.copy()
# one_hot_df[['human', 'robot']] = None
```

```python
for instance in unique_value:
    one_hot_df[instance] = (one_hot_df['whoAmI'] == instance).astype(int)

one_hot_df 
```
