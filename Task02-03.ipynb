#### Задание 2. Анализ расходов по возрасту.

Постройте линейный график, где по оси X будет отображаться возраст (age), а по оси
Y — балл по расходам (spending_score). Этот график поможет визуализировать, как
изменяются расходы в зависимости от возраста сотрудников. Проанализируйте тренды
и выявите возможные закономерности.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')
print(data.info())
data
```

```python
sns.lineplot(data=data, x='age', y='spending_score')
plt.title('Age VS Spending score')
plt.xlabel('Age')
plt.ylabel('Spending score')
```

#### Выводы
Пик расходов приходится на сотрудников среднего возраста (35 лет). Вероятно это обусловлено тем, что эта категория людей имеет семьи, повышенные комунальные расходы, расходы на содержание детей, обученеи и прочее. Сотрудники молодого возраста еще не имеют семей, не закредитованы. Пожилые мотрудники наоборот уже не несут расходы на содержание семьи, траты в основном идут на обеспечение приемлимых бытовых условий
